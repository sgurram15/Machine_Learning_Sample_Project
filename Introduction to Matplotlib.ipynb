```python
over_50 = heart_disease [heart_disease["age"] > 60]
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter",
            x="age",
            y ="chol",
            c ="fbs" );
```

```python
#OO method to plot now
fix, ax=plt.subplots(figsize=(10,6))
over_50.plot( x="age",
                y="chol",
                kind="scatter",
             c="target",
                ax=ax);
#ax.legend(*scatter.legend_elements,title="Target");
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python

fix, ax=plt.subplots(figsize=(10,5))
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                   )
ax.set(xlabel="Age",
      ylabel="Cholesterol",
      title="Heart Disease Age and Cholesterol Levels");
ax.legend(*scatter.legend_elements(),title="Target");

ax.axhline(over_50["chol"].mean(),linestyle="--");
```
